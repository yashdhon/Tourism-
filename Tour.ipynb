```python
country=data.groupby(['Continent','Country'])['Country'].count().reset_index(name='Count')
country
```

```python
africa=country[country['Continent']=='Africa']
africa=africa.sort_values(by='Count',ascending=False)
africa=africa.head()
```

```python
sns.barplot(x='Continent',y='Count',hue='Country',data=africa)
plt.title("Top Countries in Africa Contribution")
```

```python
# Get top N per continent
N = 5
top_per_continent = (
    country.sort_values(['Continent','Count'], ascending=[True,False])
           .groupby('Continent')
           .head(N)
)
```

```python
import math

continents = top_per_continent['Continent'].unique()
n_cont = len(continents)
TOP_N = 5


cols = 2  # number of columns of plots (adjust to your preference)
rows = math.ceil(n_cont / cols)


fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 5))
axes = axes.flatten()  # make it easy to index


for i, cont in enumerate(continents):
    data_cont = top_per_continent[top_per_continent['Continent'] == cont]

    sns.barplot(x='Country', y='Count', data=data_cont, palette='crest', ax=axes[i])

    axes[i].set_title(f"Top {TOP_N} Countries in {cont} by Contribution", fontsize=12)
    axes[i].set_xlabel('Country')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

for j in range(i+1, rows*cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
