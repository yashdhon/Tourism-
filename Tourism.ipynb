```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
processed['country']=le.fit_transform(processed['Country'])
processed['continent']=le.fit_transform(processed['Continent'])
processed['cityname']=le.fit_transform(processed['CityName'])
```

```python
x=processed.drop(['Country','Continent','CityName','Rating','TransactionId','UserId','AttractionTypeId','Attraction','AttractionAddress'],axis=1)
```

```python
y=processed['Rating']
```

```python
x.head()
```

```python
x.shape
```

```python
y.dropna()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
feature_names=x.columns
```

```python
lr.coef_
```

```python
feature_importance = pd.Series(
    abs(lr.coef_), index=feature_names
).sort_values(ascending=False)

print(feature_importance)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error


r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)

x_test_scaled=scaler.fit_transform(x_test)
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error


r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(
    n_estimators=500,
    random_state=42,
    n_jobs=-1
)

rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error


r2_score(y_test,y_pred)
```
